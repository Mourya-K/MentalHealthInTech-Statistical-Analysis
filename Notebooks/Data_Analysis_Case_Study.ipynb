**Gender Distribution**:

Analyze the gender distribution within the tech industry.

* **Data Visualization with Seaborn and Matplotlib:**The code uses the Seaborn and Matplotlib libraries in Python to create a bar plot (countplot) for visualizing the distribution of respondents by gender. This is a common data visualization technique used for exploring and presenting categorical data.

* **Customized Plot Appearance:** The code includes various customization options to improve the appearance of the plot. It sets a white grid background style, specifies the figure size, adds a title ("Distribution of Respondents by Gender"), and labels the x and y-axes appropriately. Additionally, it uses a specific color palette ("Set2") to style the plot.

* **Data Annotation:** The code includes a loop that iterates through the bars in the plot (represented by patches) and annotates each bar with its corresponding count. This annotation helps viewers quickly understand the exact count of respondents for each gender category. It enhances the informativeness of the visualization.

```python
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")


sns.countplot(data=df, x='Gender', order=df['Gender'].value_counts().index, palette="Set2")


plt.title('Distribution of Respondents by Gender', fontsize=16)
plt.xlabel('Gender', fontsize=14)
plt.ylabel('Count', fontsize=14)

ax = plt.gca()
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom', fontsize=12)

plt.show()
```

**Family History Distribution**:

Understand how many respondents have a family history of mental health issues.

* **Data Visualization with Seaborn and Matplotlib:** Similar to the previous code snippet, this code utilizes Seaborn and Matplotlib for data visualization. It creates a bar plot (countplot) to visualize the distribution of respondents based on their family history of mental health issues.

* **Customized Plot Appearance:** The code customizes the appearance of the plot by setting a white grid background style, specifying the figure size, and labeling the title ("Distribution of Respondents by Family History of Mental Health"), x-axis, and y-axis. It also rotates the x-axis labels to make them more readable when dealing with categorical data. The color palette "Set2" is used to style the plot.

* **Data Annotation with Percentages:** To enhance the informativeness of the plot, the code includes annotations on each bar. These annotations display both the count and the percentage of respondents for each family history category. This additional information helps viewers understand not only the count but also the relative proportion of respondents within each category.

```python
family_history_counts = df['family_history'].value_counts()
total_respondents = len(df)
family_history_percentages = (family_history_counts / total_respondents) * 100

plt.figure(figsize=(8, 6))
sns.set_style("whitegrid")

ax = sns.countplot(data=df, x='family_history', palette="Set2")


plt.title('Distribution of Respondents by Family History of Mental Health', fontsize=16)
plt.xlabel('Family History', fontsize=14)
plt.ylabel('Count', fontsize=14)


plt.xticks(rotation=0, fontsize=12)


for p, label in zip(ax.patches, family_history_counts.index):
    height = p.get_height()
    ax.annotate(f'{height} ({family_history_percentages[label]:.2f}%)', (p.get_x() + p.get_width() / 2., height),
                ha='center', va='bottom', fontsize=10, color='black')


plt.show()
```

**No. of Employees, Care Options, and Leave Countplots:**

Analyze the distribution of respondents by the number of employees, care options, and leave options.

* **Automated Countplot Generation:** The code uses a loop to generate countplots for three categorical variables: 'no_employees,' 'care_options,' and 'leave.' For each variable, a countplot is created to visualize the distribution of respondents based on their choices or responses to that specific variable.

* **Consistent Customization:** The code applies consistent customization options for each countplot. It sets a white grid background style, specifies the figure size, adds a title based on the variable being plotted (e.g., "Distribution of Respondents by no_employees"), labels the x-axis and y-axis appropriately, and rotates the x-axis labels by 45 degrees for better readability. The "Set2" color palette is used for styling.

* **Insights into Survey Responses:** Each countplot provides insights into how respondents answered questions related to 'no_employees' (company size), 'care_options' (care benefits provided by the employer), and 'leave' (types of leave offered by the employer). By visualizing the distribution of responses for each variable, viewers can gain an understanding of the preferences and choices of the survey respondents, which can be valuable for making decisions or drawing conclusions based on the survey data.

```python
countplot_variables = ['no_employees', 'care_options', 'leave']


for col in countplot_variables:

    plt.figure(figsize=(10, 6))
    sns.set_style("whitegrid")


    sns.countplot(data=df, x=col, order=df[col].value_counts().index, palette="Set2")


    plt.title(f'Distribution of Respondents by {col}', fontsize=16)
    plt.xlabel(col, fontsize=14)
    plt.ylabel('Count', fontsize=14)


    plt.xticks(rotation=45, fontsize=12)


    plt.show()
```
