```python
def map_gender(gender):
    # Normalize the input to lowercase
    gender = gender.lower()
    
    # Define mappings based on the provided counts
    if gender in ['male', 'm', 'cis male', 'man','cis man','make']:
        return 'Male'
    elif gender in ['female', 'f', 'cis female', 'cis-female/femme','woman']:
        return 'Female'
    else:
        return 'Other'  # Any other gender value is categorized as 'Other'
```

```python
df["Gender"] = df["Gender"].map(map_gender)
```

```python
df["Gender"].value_counts()
```

```python
df = df.dropna()
```

```python
df = df[df["Gender"]!="Other"]
```
